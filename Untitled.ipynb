# Description of data
A radiation detection experiment was performed using an organic detector and a Cf-252 source (which emits both gamma and neutrons). We wanted to differentiate the neutron signals from the gamma signals generated by the detector. The discrimination parameter (*DP*) was defined as the ratio of the area under the detector pulse subtracting the area  under  its  tail ($Q_S$), to  the  total  area  ($Q_L$) under the pulse ($DP=\frac{Q_S}{Q_L}$). 

The detector pulses were analyzed and their *DP* was calculated and saved in psd4.txt. This file was read below: 

```python
import numpy as np
from scipy.stats import norm

#opening a data set to 
with open(r"C:\Users\mmishra\OneDrive - North Carolina State University\bayesian_inference/psd4.txt","r") as inp:
    psd1 = []
    for line in inp:
        psd1.append(line)

psd1 = np.array(psd1)
psd1 = psd1.astype(np.float)
```

```python
#plot histogram
def hist1(ql_slice,nbins):
    hq, bnedgess  = np.histogram(ql_slice,bins=nbins)
    plt.hist(ql_slice, bins=nbins,color = "gray")
    return hq,bnedgess
    
#fit single gaussian
def gauss(x,mu,var,A):
    return A*np.exp(-(x-mu)**2/2/var)
#fit two gaussians
def bimodal(x,mu1,var1,A1,mu2,var2,A2):
    return gauss(x,mu1,var1,A1)+gauss(x,mu2,var2,A2)
```

The histogram of parameter *DP* was plotted. Since neutron pulse has a longer tail than a gamma pulse, the neutron events are concentrated around the first peak and gamma events are concentrated around the bigger peak to its right in the histogram below. The histogram was fitted with a double gaussian shown in red. 

```python
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

hq2,bne2 = hist1(psd1,800)
plt.xlim([0.5, 1.0])
plt.xlabel('$DP = \\frac{Q_{S}}{Q_{L}}$',fontsize = 16)
plt.ylabel('counts',fontsize = 16)
bne22=(bne2[1:]+bne2[:-1])/2
params,cov=curve_fit(bimodal,bne22,hq2,p0=[0.85,0.06,60.,0.69,0.07,20])
plt.plot(bne22,bimodal(bne22,*params),color='red',lw=2,label='model')
```

Bayesian inference was applied to assign *DP* to a neutron class ($C_n$) or a gamma class ($C_g$). The intersection of two gaussian was calculated, which maximized the posterior probability and divided the *DP* into $C_n$ and $C_g$. The values to the left of the dotted line (DP = 0.795) are assigned to $C_n$, while the values to the right of the dotted line are assigned to $C_g.$

```python
xcx = np.arange(0.5,1.,0.0001)
ycx1 = gauss(xcx,params[0], params[1], params[2])
ycx2 = gauss(xcx,params[3],params[4],params[5])
plt.plot(xcx, bimodal(xcx,*params))
idx = np.argwhere(np.diff(np.sign(ycx1 - ycx2))).flatten()
plt.plot(xcx[idx[0]], ycx1[idx[0]], 'ro')
plt.ylabel('counts',fontsize = 16)
plt.xlabel('$DP = \\frac{Q_{S}}{Q_{L}}$',fontsize = 16)
plt.axvline(x=xcx[idx[0]],linestyle='--',color='black')
plt.show()
```

The posterior probabilities p($C_g$|DP) and p($C_n$|DP) are plotted below. The dotted line DP = 0.795 gives the decision boundary which gives the minimum misclassification rate.
